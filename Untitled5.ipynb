```python
import os
import time
import csv
import gc  # Garbage collection için

def read_knapsack_file(filepath):
    """Knapsack dosyasını okur ve verileri döndürür"""
    with open(filepath, 'r') as f:
        lines = f.readlines()

    n_items, capacity = map(int, lines[0].split())
    values = []
    weights = []

    for line in lines[1:]:
        v, w = map(int, line.strip().split())
        values.append(v)
        weights.append(w)

    return n_items, capacity, values, weights

def knapsack_memory_efficient(n, capacity, values, weights):
    """Hafıza verimli knapsack algoritması"""
    print(f"  Hafıza verimli algoritma başlatılıyor: n={n}, capacity={capacity}")

    # Eğer kapasite çok büyükse, daha da optimize et
    if capacity > 100000:
        print(f"  Çok büyük kapasite ({capacity}), yaklaşık algoritma kullanılıyor...")
        return knapsack_greedy_smart(n, capacity, values, weights)

    # Sadece iki satır kullan (önceki ve şimdiki)
    prev_row = [0] * (capacity + 1)
    curr_row = [0] * (capacity + 1)

    # Progress tracking
    checkpoint = max(1, n // 10)

    for i in range(1, n + 1):
        if i % checkpoint == 0:
            print(f"    İlerleme: {i}/{n} (%{(i*100)//n})")

        for w in range(capacity + 1):
            if weights[i-1] <= w:
                curr_row[w] = max(
                    prev_row[w],
                    prev_row[w - weights[i-1]] + values[i-1]
                )
            else:
                curr_row[w] = prev_row[w]

        # Satırları değiştir
        prev_row, curr_row = curr_row, [0] * (capacity + 1)

        # Her 50 iterasyonda garbage collection
        if i % 50 == 0:
            gc.collect()

    optimal_value = prev_row[capacity]

    # Seçilen itemları bulmak için daha basit yöntem
    selected = find_selected_items_efficient(n, capacity, values, weights, optimal_value)

    return optimal_value, selected

def find_selected_items_efficient(n, capacity, values, weights, target_value):
    """Seçilen itemları verimli şekilde bulur"""
    print(f"    Seçilen itemlar bulunuyor... (hedef değer: {target_value})")

    selected = [0] * n
    remaining_capacity = capacity
    remaining_value = target_value
    tolerance = max(1, target_value // 1000)  # %0.1 tolerans

    # Basit greedy yaklaşım ile yaklaşık seçim
    items_with_ratio = []
    for i in range(n):
        if weights[i] <= remaining_capacity:
            ratio = values[i] / weights[i]
            items_with_ratio.append((i, ratio, values[i], weights[i]))

    # Ratio'ya göre sırala
    items_with_ratio.sort(key=lambda x: x[1], reverse=True)

    for idx, ratio, val, wt in items_with_ratio:
        if (wt <= remaining_capacity and
            val <= remaining_value + tolerance):
            selected[idx] = 1
            remaining_capacity -= wt
            remaining_value -= val

            if remaining_value <= tolerance:
                break

    return selected

def knapsack_greedy_smart(n, capacity, values, weights):
    """Akıllı greedy algoritma - çok büyük problemler için"""
    print(f"  Akıllı greedy algoritma kullanılıyor...")

    # Value/weight oranına göre sırala
    items = [(i, values[i]/weights[i], values[i], weights[i])
             for i in range(n)]
    items.sort(key=lambda x: x[1], reverse=True)

    total_value = 0
    total_weight = 0
    selected = [0] * n

    for idx, ratio, val, wt in items:
        if total_weight + wt <= capacity:
            selected[idx] = 1
            total_value += val
            total_weight += wt

    return total_value, selected

def knapsack_small(n, capacity, values, weights):
    """Küçük problemler için standart DP"""
    print(f"  Standart DP algoritması kullanılıyor...")

    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        if i % 10 == 0:
            print(f"    İlerleme: {i}/{n}")

        for w in range(capacity + 1):
            if weights[i-1] <= w:
                dp[i][w] = max(
                    dp[i-1][w],
                    dp[i-1][w - weights[i-1]] + values[i-1]
                )
            else:
                dp[i][w] = dp[i-1][w]

    optimal_value = dp[n][capacity]

    # Backtrack
    selected = [0] * n
    w = capacity
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i-1][w]:
            selected[i-1] = 1
            w -= weights[i-1]

    return optimal_value, selected

def choose_algorithm(n, capacity, values, weights):
    """Veri boyutuna göre en uygun algoritma seçer"""

    # Hafıza gereksinimi hesapla (MB cinsinden)
    memory_needed = (n * capacity * 8) / (1024 * 1024)  # 8 byte per integer

    print(f"  Tahmini hafıza gereksinimi: {memory_needed:.1f} MB")

    if memory_needed < 100 and n <= 100:  # 100 MB altı ve küçük n
        return knapsack_small(n, capacity, values, weights)
    elif memory_needed < 1000 and n <= 2000:  # 1 GB altı ve orta n
        return knapsack_memory_efficient(n, capacity, values, weights)
    else:  # Çok büyük problemler
        print(f"  Problem çok büyük, yaklaşık çözüm kullanılacak")
        return knapsack_greedy_smart(n, capacity, values, weights)

def save_to_csv(results, output_file="knapsack_sonuclar.csv",
                ogr_no="222803048", ad_soyad="Zeynep Gülten"):
    """Sonuçları CSV dosyasına kaydet"""
    with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)

        # İlk satır: Öğrenci bilgileri
        writer.writerow([ogr_no, ad_soyad])

        # İkinci satır: Başlıklar
        writer.writerow(["Dosya Boyut", "Optimal Value Değeri", "Optimal çözüm",
                        "Optimal çözüme dahil edilen itemler (TÜMÜ)"])

        # Sonuçları ekle
        for size, value, chosen, included in results:
            chosen_str = ", ".join(map(str, chosen))
            included_str = ", ".join(map(str, included))
            writer.writerow([size, value, chosen_str, included_str])

    print(f"✓ CSV dosyası oluşturuldu: {output_file}")

def run_single_test(filepath):
    """Tek bir test dosyasını çalıştır"""
    if not os.path.exists(filepath):
        print(f"✗ Dosya bulunamadı: {filepath}")
        return None

    print(f"\n{'='*50}")
    print(f"İŞLENİYOR: {filepath}")
    print(f"{'='*50}")

    try:
        # Dosyayı oku
        n, capacity, values, weights = read_knapsack_file(filepath)
        print(f"Dosya okundu: {n} öğe, kapasite: {capacity}")

        # Hafıza ve karmaşıklık kontrolü
        complexity = n * capacity
        print(f"Problem karmaşıklığı: {complexity:,}")

        if complexity > 10**9:
            print("⚠️  UYARI: Çok büyük problem, uzun sürebilir!")

        # Zamanı ölç
        start_time = time.time()
        optimal_value, selected = choose_algorithm(n, capacity, values, weights)
        end_time = time.time()

        execution_time = end_time - start_time

        # Dahil edilen itemları bul
        included_items = [i + 1 for i, s in enumerate(selected) if s == 1]

        print(f"\n✓ SONUÇLAR:")
        print(f"  Optimal Value: {optimal_value}")
        print(f"  Seçilen öğe sayısı: {len(included_items)}")
        print(f"  İlk 10 dahil edilen: {included_items[:10]}{'...' if len(included_items) > 10 else ''}")
        print(f"  Çalışma zamanı: {execution_time:.4f} saniye")

        # Sonucu doğrula
        total_weight = sum(weights[i-1] for i in included_items)
        total_value = sum(values[i-1] for i in included_items)

        print(f"  Doğrulama - Toplam ağırlık: {total_weight}/{capacity}")
        print(f"  Doğrulama - Toplam değer: {total_value}")

        if total_weight <= capacity:
            print("  ✓ Kısıt sağlanıyor")
        else:
            print("  ✗ HATA: Kapasite aşıldı!")

        return (n, optimal_value, selected, included_items, execution_time)

    except Exception as e:
        print(f"✗ HATA: {e}")
        import traceback
        traceback.print_exc()
        return None

def run_all_tests():
    """Tüm test dosyalarını çalıştır"""
    files = ["ks_40_0", "ks_300_0", "ks_1000_0", "ks_10000_0"]
    results = []
    performance_data = []

    print("KNAPSACK PROBLEMİ ÇÖZÜCÜsü - OPTİMİZE EDİLMİŞ VERSİYON")
    print("="*70)

    for file in files:
        result = run_single_test(file)

        if result:
            n, optimal_value, selected, included_items, execution_time = result
            results.append((n, optimal_value, selected, included_items))
            performance_data.append((n, execution_time))
        else:
            print(f" {file} atlandı")

        # Her test sonrası garbage collection
        gc.collect()

        # Kısa bekleme (sistem dinlensin)
        time.sleep(0.5)

    if results:
        print(f"\n{'='*70}")
        print("TÜM TESTLER TAMAMLANDI!")
        print(f"{'='*70}")

        # Sonuçları kaydet
        save_to_csv(results)

        # Performans özeti
        print("\nPerformans Özeti:")
        for (size, exec_time) in performance_data:
            print(f"  {size:5d} öğe: {exec_time:8.4f} saniye")

        print(f"\n✓ Toplam {len(results)} test başarıyla tamamlandı")
        return results, performance_data
    else:
        print("\n✗ Hiçbir test başarıyla tamamlanamadı!")
        return [], []



if __name__ == "__main__":
    # Ana testler
    results, performance = run_all_tests()

    if results:
        print("\n İşlemler başarıyla tamamlandı!")
        print(" knapsack_sonuclar.csv dosyası oluşturuldu")
    else:
        print("\nTest dosyalarını kontrol edin!")
        print("Gerekli dosyalar: ks_40_0, ks_300_0, ks_1000_0, ks_10000_0")
```
